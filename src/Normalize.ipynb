```python
import pandas as pd
import sqlite3
from collections import Counter
import numpy as np
```

```python
def read_episodes_from_file(filename):
    Episodes = []
    Columns = []
    with open(filename, 'r') as f:
        for line in f:
            if "Begin-----" in line:
                Episode = []
            elif "End-----" in line:
                Episodes.append(Episode)
                Episode = []
            else:
                temp = line.split(': ')[-1].strip()
                Episode_type = line.split(': ')[0].strip()
                if Episode_type not in Columns:
                    Columns.append(Episode_type)
                Episode.append(temp)
    return Episodes,Columns

def modify_features (line):
    features = []
    line = line[1:-1].replace("'","").replace(" ","").split(",")
    for feature in line:
        features.append(feature)
    return features
```

```python
episodes,Columns = read_episodes_from_file("Dataload_UW_dataset.log")

data_len_3 = []
data_len_6 = []
for data in episodes:
    if len(data) == 3:
        data_len_3.append(data)
    else:
        data_len_6.append(data)
        
df_data_len_6 = pd.DataFrame(data_len_6)
df_data_len_6.columns = Columns

df_data_len_3 = pd.DataFrame(data_len_3)
df_data_len_3.columns = [Columns[0], Columns[-2], Columns[-1]]
len(df_data_len_3)
```
